```python
#BNB-Binance ,ADA-Cardano,DOT-Polkadot,''
cryptocurrency =['BTC-INR','ETH-INR','BNB-INR','SOL-INR','ADA-INR','DOT-INR','MATIC-INR','BAT-INR','AAVE-INR','MANA-INR','SAND-INR','AXS-INR']
for crypto in cryptocurrency:
    start = dt.datetime(2020,11,11)
    end = dt.datetime.now()
    crypto_data = yf.download(crypto, start, end)
    plt.figure(figsize=(16,5))
    plt.title(crypto)
    sns.lineplot(y = 'Adj Close',x = 'Date',data = crypto_data)
    plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
    plt.show()
```
