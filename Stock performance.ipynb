```python
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

## Marcellous -->Consistant compounder

```python
# CONSISTANT COMPOUNDER SM
ccp = ['ABBOTINDIA.NS','ASIANPAINT.NS','BERGEPAINT.NS','NESTLEIND.NS','PIDILITIND.NS','TITAN.NS','DIVISLAB.NS','TCS.NS']
for stk in ccp:
    start = dt.datetime(2020,1,1)
    end = dt.datetime.now()
    stk_data = yf.download(stk, start, end)
    plt.figure(figsize=(16,5))
    plt.title(stk)
    sns.lineplot(y = 'Adj Close',x = 'Date',data = stk_data)
    plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
    plt.show()
```
